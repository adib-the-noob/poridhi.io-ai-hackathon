```python
import chromadb
from chromadb.config import Settings

# Set up local storage path for ChromaDB
# local_storage_path = '/content/chromadb_storage'

# Initialize ChromaDB client with the local directory for storing data
client = chromadb.HttpClient(host="54.169.92.49", port="8000")
```

```python
import json

# Load product data


with open('products.json', 'r') as f:
    products = json.load(f)
    print(products)

# Create or get an existing collection
collection = client.get_or_create_collection(name="product_collection")

# Add products to ChromaDB collection
for product in products:
    product_id = product.get("id")
    title = product.get("title", "")
    description = product.get("description", "")

    # Ensure none of the fields are None
    if title is None: title = ""
    if description is None: description = ""

    # Combine title and description for better search context
    document_text = title
    if description:
        document_text += " - " + description

    collection.add(
        ids=[str(product_id)],
        documents=[document_text],
        metadatas=[{"product_id": product_id, "title": title}],
    )
```

```python
# prompt: load all the data from database, n a list

# Assuming client and collection are already initialized as in the previous code
# ... (previous code to initialize client and collection)

all_data = []

# Get all documents from the collection
results = collection.get(include=["documents", "metadatas", "ids"])

# Iterate through the results and append the relevant information to the list
for i in range(len(results["ids"])):
    data_point = {
        "id": results["ids"][i],
        "document": results["documents"][i],
        "metadata": results["metadatas"][i]
    }
    all_data.append(data_point)

# Now all_data holds all the documents, metadatas, and ids from the collection
all_data
```
